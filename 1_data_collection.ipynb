```python
# API endpoint to get channel details
url = f'https://www.googleapis.com/youtube/v3/channels?part=snippet,statistics&id={channel_id}&key={api_key}'

# Send GET request to fetch channel data
response = requests.get(url)
channel_data = response.json()
```

```python
# Step 1: Get the playlist ID of the channel's uploaded videos
url = f'https://www.googleapis.com/youtube/v3/channels?part=contentDetails&id={channel_id}&key={api_key}'

# Send request to get playlist ID
response = requests.get(url)
channel_data = response.json()

# Extract the uploads playlist ID
uploads_playlist_id = channel_data['items'][0]['contentDetails']['relatedPlaylists']['uploads']
```

```python
# Step 2: Initialize variables to fetch video information
videos_fetched = 0
max_videos = 5000
next_page_token = None

# Loop through pages until we fetch the desired number of videos or there are no more pages
while videos_fetched < max_videos:
    # Step 3: Fetch video IDs from the uploads playlist
    url_videos = f'https://www.googleapis.com/youtube/v3/playlistItems?part=snippet&playlistId={uploads_playlist_id}&maxResults=50&key={api_key}'

    # If there is a nextPageToken, include it in the request to get the next set of videos
    if next_page_token:
        url_videos += f"&pageToken={next_page_token}"

    response_videos = requests.get(url_videos)
    videos_data = response_videos.json()

    # If there are no more videos- break out of the loop
    if 'items' not in videos_data:
        print("No more videos found.")
        break

    # Process each video in the response
    for item in videos_data['items']:
        video_id = item['snippet']['resourceId']['videoId']
        video_title = item['snippet']['title']
        video_description = item['snippet']['description']
        publish_time = item['snippet']['publishedAt']

        # Fetch video statistics
        video_details_url = f'https://www.googleapis.com/youtube/v3/videos?part=statistics&id={video_id}&key={api_key}'
        video_details_response = requests.get(video_details_url)
        video_stats = video_details_response.json()


        # Check if 'statistics' field exists and has the necessary keys
        if 'items' in video_stats and len(video_stats['items']) > 0:
            statistics = video_stats['items'][0].get('statistics', {})
            view_count = statistics.get('viewCount', 'Not Available')
            like_count = statistics.get('likeCount', 'Not Available')
            comment_count = statistics.get('commentCount', 'Not Available')
        else:
            # Handle missing video statistics
            view_count = 'Not Available'
            like_count = 'Not Available'
            comment_count = 'Not Available'

        # Print video details
        print(f"Title: {video_title}")
        print(f"Published at: {publish_time}")
        print(f"Description: {video_description}")
        print(f"Views: {view_count}")
        print(f"Likes: {like_count}")
        print(f"Comments: {comment_count}")
        print('-' * 50)

        # Increment the count of videos fetched
        videos_fetched += 1

        # Break if weâ€™ve fetched enough videos
        if videos_fetched >= max_videos:
            break

    # Step 4: Check if there's a nextPageToken for more videos
    next_page_token = videos_data.get('nextPageToken', None)
    if not next_page_token:
        break

print(f"Total videos fetched: {videos_fetched}")
```
